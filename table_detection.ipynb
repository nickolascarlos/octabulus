```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread(r"./sample.png", 1)
```

```python
def shw(img):
    plt.title('Imagem')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
_, mask = cv2.threshold(image, 233, 250, cv2.THRESH_BINARY_INV)
```

```python
shw(mask)
```

```python
vert_m = cv2.getStructuringElement(cv2.MORPH_RECT, (1, image.shape[0]//20))
```

```python
res_vert = cv2.dilate(image, vert_m)
```

```python
res_vert = cv2.erode(res_vert, vert_m)
```

```python
res_vert = cv2.bitwise_not(res_vert)
```

```python
shw(res_vert)
```

```python
hor_m = cv2.getStructuringElement(cv2.MORPH_RECT, (image.shape[1]//15, 1))
```

```python
res_hor = cv2.dilate(image, hor_m)
```

```python
res_hor = cv2.erode(res_hor, hor_m)
```

```python
res_hor = cv2.bitwise_not(res_hor)
```

```python
shw(res_hor)
```

```python
cbon = cv2.bitwise_or(res_hor, res_vert)
```

```python
shw(cbon)
```

```python
blurred_cbon = cv2.blur(cbon, (3,3))
```

```python
shw(blurred_cbon)
```

```python
_, blurred_thresh_cbon = cv2.threshold(blurred_cbon, 30, 250, cv2.THRESH_BINARY)
```

```python
shw(blurred_thresh_cbon)
```

```python
contours, hierarchy = cv2.findContours(blurred_thresh_cbon, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
shw(cv2.drawContours(np.copy(image), contours, 10, (0, 0, 255), 3))
```

```python
contours
```

```python
rct = cv2.boundingRect(contours[0])
```

```python
shw(cv2.rectangle(np.copy(image), (rct[0], rct[1]), (rct[0]+rct[2], rct[1]+rct[3]) , (0,255,0), 3))
```
